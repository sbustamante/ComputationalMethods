```python
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt
```

```python
#puntos escogidos
X=[0,2.5,3.1]
Y=[0,2.4,1]
#polinomio calculado en clase
x_pol=np.linspace(0,3.1)
pol_aprox=np.poly1d([-1,3.45,0])
#polinomio usando scipy
pol_sci=interpolate.interp1d(X,Y,kind='quadratic')
#graficando puntos
plt.plot(X,Y,'bo')
#graficando polinomio aproximado
plt.plot(x_pol,pol_aprox(x_pol),label='Polinomio aproximado en clase: -x^2+3.5x')
#graficando polinomio de scipy
plt.plot(x_pol,pol_sci(x_pol),label='Polinomio con scipy')
plt.legend()
plt.show()
```
