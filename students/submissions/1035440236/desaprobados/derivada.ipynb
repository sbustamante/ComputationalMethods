```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
def f(x,a=1):   # Definimos nuestra funcion
  return a*np.sin(x)   # Tenemos nuestra funcion seni
c=derivative(f, 1.0, dx=1e-6,n=1)   # La derivamos.
print(c)


# Procedemos a graficar.
# Grafiquemos nuestra funcion
x=np.linspace(-np.pi,np.pi)
plt.plot(x,f(x),label="Funcion original",lw=4)
y = derivative(f,1,dx=1e-6)*(x - 1) + f(1)
plt.plot(x,y,label="Derivada en un punto.")
plt.plot(1,f(1),"*y",label="Punto")
plt.grid()
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Grafico")
plt.show()
```

```python
# Tenemos la derivada anlitica:
def derivada_analitica (x,a=1):
  return a*np.cos(x)

derivada_analitica(1)

# Realizemos una resta
Diferencia=c-derivada_analitica(1)
print("la diferencia es",Diferencia)
 # Podemos observar que es muy peque√±o y corresponde a un cero numerico.


```
