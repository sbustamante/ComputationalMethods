```python
#Libraries
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
def integrand(x,k):
    return 1/np.sqrt(1- (k**2)*(np.sin(x))**2)

def Period(K, l, g):
    return 4*np.sqrt(l/g)*K
   

l = 0.2 
g = 9.81
deg = np.array([0,10,20,30,40,50,60])
θ = deg * np.pi / 180 #degrees to radian
k = np.sin(θ)/2


T = []

for i in range(len(k)):
    I, err = integrate.quad(integrand, 0, 2*np.pi, args =(k[i]))
    T.append(Period(I, l, g))
    

#Plot
plt.figure(figsize=(10, 6), dpi=80)
plt.plot(deg, T, color = "green")
plt.ylabel("Period $(T)$")
plt.xlabel("Angle $(θ)$")
plt.title(" T vs θ")
plt.grid()
```
