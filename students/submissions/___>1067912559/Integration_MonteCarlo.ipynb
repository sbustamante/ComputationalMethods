## Homework 5 : Integration Monte Carlo Method 

```python
#Libraries
import numpy as np
import scipy.integrate as integrate
import matplotlib.pyplot as plt
```

```python
def f(x):
    #Integrand
    return np.sqrt(15*(x**3) +21* (x**2) +14*x +3)* np.exp(-0.5 *x)

def IMC(f, a, b, N):
    #Mean Value Method
    x = np.random.uniform(a, b, N)
    y = f(x)
    return (b-a)*np.mean(y)

a = 0
b = 44
N = [10, 100, 1000, 10000, 100000, 1000000]

#Value of the integral using scipy.integrate.quad 
I, error = integrate.quad(f, a, b)

vIMC = np.vectorize(IMC)
Imc = vIMC(f, a, b, N) 

#PLOT
plt.figure(figsize=(10, 6))
plt.xscale("log")
plt.scatter(N, Imc, color = "red", label = "Integrals Mean Value Method")
plt.axhline(I, xmin=0, xmax=1e6, label = "Scipy.integrate.quad")
plt.xlabel("Number of iterations")
plt.ylabel("Value of the integral")
plt.legend()
plt.grid();
```
