```python
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt

def fun(x,a):
    return a*np.sin(x)
```

```python
xmin = 0.0
xmax = 20.0
x=np.linspace(xmin, xmax, 100)

def Derivada(fun,xmin,xmax,a):     #Como Derivative halla la derivada en un sólo punto, se debe hacer una función
    x=np.linspace(xmin, xmax, 100) #para que halle las diferentes derivadas en todos los puntos del rango dado.
    Der=[]
    aa={a}
    for i in range (0,len(x)):
        k=x[i]
        Dp=derivative(fun, k, dx=1e-4,args=(aa))    #Derivative acepta los argumentos en tuplas, por eso aa={}
        Der.append(Dp)
    return Der

plt.plot(x, fun(x,1), color='y', label='1*sin(X)')            #función con 
plt.plot(x, Derivada(fun,xmin,xmax,1), '-m', label='Derivada con a=1')

plt.grid()
plt.legend()
```

```python
for i in range(1,6):
    color=[a,'b','c','m','g','y','k','r']
    plt.plot(x, fun(x,i), color=color[i], label='{}*sin(X)'.format(i))
    plt.plot(x, Derivada(fun,xmin,xmax,i), color=color[-i], label='Derivada con a={}'.format(i))
    plt.title('Gráfica de la función con a=1, 2, 3, 4 y 5 y sus derivadas')
    plt.grid()
    plt.legend()
    
```
